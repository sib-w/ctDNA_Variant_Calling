```python
## Subset SRA Run Table to select patient to analyze
## FFPE = formalin-fixed, paraffin-embedded tissue (baseline tumor)
## PBMC = Peripheral blood mononuclear cell (filter out germline mutations)
## CTC = ctDNA
## I selected patients who have all three sample types (FFPE, PBMC, and CTC)
##to ensure all possible comparison combinations are available.

import pandas as pd

df = pd.read_csv("SraRunTable.csv") # Load file

# Subset CTC samples and Extract patient ID (CTC### part)
ctc = df[df['Sample Name'].str.contains("CTC", case=False, na=False)]

ctc['patient_id'] = ctc['Sample Name'].str.extract(r"(CTC\d+)")

patients = ctc.groupby('patient_id') # Group by patient

# Show each patient and their runs
for pid, group in patients:
    print(f"\n=== Patient {pid} ===")
    display(group[['Run','Sample Name','Library Name','BioSample']])

# Patient Selection Criteria
## Minimum of two time points for CTC measurements
## PBMC data available to serve as normal/control

#=== Patient CTC364 === SELECTED PATIENT
#	      Run		    Sample Name	  Library Name	BioSample
#166	SRR13973869	PBMC_CTC364	  PBMC_76		    SAMN18317964
#174	SRR13973877	FFPE_CTC364	  colorectal_70	SAMN18318305
#416	SRR13974119	CTC364		    plasma_271	  SAMN18318235
#417	SRR13974120	CTC364-2	    plasma_270	  SAMN18318234
```
